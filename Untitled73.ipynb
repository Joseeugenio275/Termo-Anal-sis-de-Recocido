```python
import numpy as np
import matplotlib.pyplot as plt

# =======================================================
# 1. PARÁMETROS FÍSICOS Y GEOMÉTRICOS (Paper)
# =======================================================
# Geometría
Diam_ext = 1.867      # m
Diam_int = 0.508      # m
Width = 1.250         # m (Altura de UNA bobina)
NUM_COILS = 4         # 4 Bobinas
H_total = Width * NUM_COILS

# Mallado (FDM)
M = 20    # Nodos radiales (Suficiente para precisión visual)
N_coil = 30 # Nodos axiales POR BOBINA
N = N_coil * NUM_COILS
dr = (Diam_ext - Diam_int) / 2 / (M - 1)
dz = Width / (N_coil - 1)
r_coords = np.linspace(Diam_int/2, Diam_ext/2, M)

# Paso de tiempo (Estabilidad)
dt = 2.0 # Segundos

# Propiedades del Acero (SPCC)
rho = 7850.0          # kg/m³
c_p = 500.0           # J/(kg K)
lambda_z = 45.0       # W/(m K) Axial (Constante)
lambda_s = 45.0       # W/(m K) Base para radial

# Parámetros para Lambda_r (Resistencia Radial - Eqs 9-17)
S_strip = 0.00150; b_o = 10.0e-6; lambda_o = 1.5; epsilon = 0.08554
sigma_sb = 5.6703e-8; P_mpa = 8.0; H_mpa = 1133.86
sigma_p = 3.2215e-6; tan_theta = 0.1

# Gas y Flujo (H2)
u_gas_base = 15.0 # m/s
T_gas_base = 300.0

# --- DATOS DE LA TABLA 1 (DISTRIBUCIÓN DE FLUJO) ---
# Factores de flujo relativo para cada bobina (1=Abajo, 4=Arriba)
# Radial: Flujo que pasa por el diámetro interior/exterior
factors_flow_radial = [1.0, 0.798, 0.652, 0.543]
# Axial: Flujo que pasa por la PLACA CONVECTORA encima de la bobina
factors_flow_axial_top = [0.202, 0.183, 0.167, 0.543]

# Tiempos (Perfil Trapezoidal)
t_subida = 5.0 * 3600
t_mant = 12.5 * 3600
t_bajada = 10.0 * 3600
total_steps = int((t_subida + t_mant + t_bajada) / dt)
TIEMPOS_OBJETIVO_H = [5.0, 15.0, 20.0, 25.0]

# =======================================================
# 2. FUNCIONES AUXILIARES
# =======================================================
def props_gas_H2(T):
    T = np.maximum(T, 100.0) # Protección
    lam_g = 0.18 + 3e-4 * (T - 300)
    mu_g = 8.9e-6 + 2e-8 * (T - 300)
    return lam_g, mu_g

def calc_h_dittus(T_gas, u_gas, flow_ratio):
    # Ec. 7: Nu = 0.023 * Re^0.8 * Pr^0.4
    # Si el flujo baja, la velocidad baja proporcionalmente.
    # h es propocional a u^0.8 -> h_eff = h_base * (ratio)^0.8
    if flow_ratio <= 0.001: return 0.0 # Sin flujo

    lam_g, mu_g = props_gas_H2(T_gas)
    rho_g = 0.08 * (300/T_gas)
    De = 0.1 # Diámetro hidráulico equivalente

    u_eff = u_gas * flow_ratio
    Re = (rho_g * u_eff * De) / mu_g
    Re = np.maximum(Re, 100.0)

    Nu = 0.023 * (Re**0.8) * (0.7**0.4)
    h = (Nu * lam_g) / De
    return h

def calc_lambda_r(T_node):
    # Modelo completo de resistencia radial (Eq 17 del paper)
    P = P_mpa * 1e6; H_ = H_mpa * 1e6
    phi = P / (H_ + P)
    b = 42.7e-6 * np.exp(-5e-2 * P_mpa)
    lam_g, _ = props_gas_H2(T_node)

    R_s2 = S_strip/(2*lambda_s)
    R_o = b_o/lambda_o
    T_k = np.maximum(T_node, 100.0)

    # Radiación (Eq 11)
    R_r = (2-epsilon)/(4*(1-phi)*epsilon*sigma_sb*(T_k**3))
    # Gas (Eq 12)
    R_g = b/((1-phi)*lam_g)
    # Contacto (Eq 13)
    R_d = (sigma_p/(1.13*lambda_s*tan_theta))*(phi**-0.94)

    # Paralelo
    R_gap = 1.0/((1/R_r)+(1/R_g)+(1/R_d))

    # Total
    R_total = 2*R_s2 + 2*R_o + R_gap
    return S_strip / R_total

def get_T_furnace(t_sec):
    # Perfil Trapezoidal
    t_h = t_sec / 3600.0
    T_start = 300.0; T_max = 983.0; T_final = 300.0

    if t_sec <= t_subida:
        return T_start + (T_max - T_start)*(t_sec/t_subida)
    elif t_sec <= (t_subida + t_mant):
        return T_max
    else:
        dt_cool = t_sec - (t_subida + t_mant)
        return T_max - (T_max - T_final)*(dt_cool/t_bajada)

# =======================================================
# 3. SIMULACIÓN PRINCIPAL
# =======================================================
T = np.full((M, N), 300.0)
T_new = T.copy()
mapas_guardados = {}

print(f"Iniciando simulación COIL-BY-COIL...")

# Pre-calcular constantes de malla
C1 = dt / (rho * c_p)
idx_report = total_steps // 10

for step in range(total_steps):
    t_curr = step * dt
    T_gas_furnace = get_T_furnace(t_curr)

    # --- MODELO DE GAS SIMPLIFICADO (Cascada) ---
    # El gas entra por abajo y se va calentando/enfriando al subir
    # Calculamos un T_gas promedio PARA CADA BOBINA
    T_gas_coils = np.zeros(NUM_COILS)
    tg_curr = T_gas_furnace
    for k in range(NUM_COILS):
        # El gas pierde/gana temperatura al pasar por la bobina anterior
        # Simplificación: factor de decaimiento leve (0.05) para simular enfriamiento del gas
        T_gas_coils[k] = tg_curr
        # Actualizar para la siguiente (el gas se "gasta" térmicamente un poco)
        # En calentamiento: el gas se enfría. En enfriamiento: el gas se calienta.
        T_surf = np.mean(T[:, k*N_coil:(k+1)*N_coil])
        tg_curr = tg_curr - 0.05 * (tg_curr - T_surf)

    # --- CÁLCULO DE PROPIEDADES (Lento pero necesario) ---
    # Lambda_r variable en todo el campo
    L_r_field = np.zeros((M, N))
    for i in range(M):
        for j in range(N):
            L_r_field[i,j] = calc_lambda_r(T[i,j])

    # --- SOLUCIÓN FDM (Nodos internos) ---
    for i in range(1, M-1):
        r_i = r_coords[i]
        # Términos comunes
        term_common = C1 * L_r_field[i,:] # Vectorizado en j para velocidad

        # Derivadas radiales (Vectorizado)
        dT_dr2 = (T[i+1,:] - 2*T[i,:] + T[i-1,:]) / dr**2
        dT_dr  = (T[i+1,:] - T[i-1,:]) / (2*dr)
        radial_part = term_common * (dT_dr2 + (1/r_i)*dT_dr)

        # Derivada axial (Vectorizado)
        # Nota: Lambda_z es cte
        dT_dz2 = (T[i, 2:] - 2*T[i, 1:-1] + T[i, :-2]) / dz**2
        axial_part = (C1 * lambda_z) * dT_dz2

        # Aplicar (Cuidado con índices de axial_part que es más corto)
        T_new[i, 1:-1] = T[i, 1:-1] + radial_part[1:-1] + axial_part

    # --- CONDICIONES DE CONTORNO (BOBINA POR BOBINA) ---
    for k in range(NUM_COILS):
        j_start = k * N_coil
        j_end   = (k + 1) * N_coil

        Tg_local = T_gas_coils[k]

        # 1. RADIAL (Interior y Exterior)
        # Usamos el factor de flujo RADIAL de la Tabla 1 para esta bobina
        h_rad = calc_h_dittus(Tg_local, u_gas_base, factors_flow_radial[k])

        # Borde Radial Interno (i=0) y Externo (i=M-1) para los nodos j de esta bobina
        for j in range(j_start, j_end):
            # Interno
            Bi_in = h_rad * dr / L_r_field[1, j]
            T_new[0, j] = (T[1, j] + Bi_in * Tg_local) / (1 + Bi_in)
            # Externo
            Bi_out = h_rad * dr / L_r_field[M-1, j]
            T_new[M-1, j] = (T[M-2, j] + Bi_out * Tg_local) / (1 + Bi_out)

        # 2. AXIAL (Placas Convectoras / Tapas)
        # Aquí está la clave: TRATAR TAPAS INDIVIDUALES CON CONVECCIÓN

        # Tapa Inferior de la Bobina k (j_start)
        # Si k=0, usa flujo base. Si k>0, usa el flujo de la placa anterior (top de k-1)
        ratio_bot = factors_flow_axial_top[k-1] if k > 0 else 0.2
        h_bot = calc_h_dittus(Tg_local, u_gas_base, ratio_bot)
        Bi_bot = h_bot * dz / lambda_z

        # Tapa Superior de la Bobina k (j_end - 1)
        ratio_top = factors_flow_axial_top[k]
        h_top = calc_h_dittus(Tg_local, u_gas_base, ratio_top)
        Bi_top = h_top * dz / lambda_z

        # Aplicar BC axial a todos los radios i
        for i in range(M):
            # Nodo inferior de esta bobina
            T_new[i, j_start] = (T[i, j_start+1] + Bi_bot * Tg_local) / (1 + Bi_bot)
            # Nodo superior de esta bobina
            T_new[i, j_end-1] = (T[i, j_end-2] + Bi_top * Tg_local) / (1 + Bi_top)

    # Actualizar campo
    T = T_new.copy()

    # Guardar frames
    t_h = t_curr / 3600.0
    for t_obj in TIEMPOS_OBJETIVO_H:
        if abs(t_h - t_obj) < (dt/3600.0 * 1.5) and t_obj not in mapas_guardados:
            print(f"Capturado mapa a {t_h:.2f} h")
            mapas_guardados[t_obj] = T.copy()

# =======================================================
# 4. VISUALIZACIÓN EXACTA
# =======================================================
if mapas_guardados:
    fig, axes = plt.subplots(1, 4, figsize=(16, 10), constrained_layout=True)

    for idx, t_val in enumerate(TIEMPOS_OBJETIVO_H):
        if t_val in mapas_guardados:
            ax = axes[idx]
            data_C = mapas_guardados[t_val] - 273.15

            # Rotar para que Y sea altura
            plot_data = np.flipud(np.rot90(data_C))

            # Auto-escala inteligente para contraste
            vmin, vmax = np.min(data_C), np.max(data_C)

            im = ax.imshow(plot_data, cmap='jet', aspect='auto', interpolation='bicubic',
                           extent=[0, 1, 0, NUM_COILS], vmin=vmin, vmax=vmax)

            # Dibujar las placas convectoras (líneas blancas)
            for k in range(1, NUM_COILS):
                ax.axhline(y=k, color='white', linewidth=2)

            ax.set_title(f'{t_val} h', fontsize=14)
            ax.set_xticks([])
            ax.set_yticks([])

            # Barra de color abajo
            cbar = plt.colorbar(im, ax=ax, orientation='horizontal', pad=0.02, aspect=30)
            cbar.set_label('Temp (°C)', fontsize=9)

    plt.suptitle("Evolución Térmica (Replicación Exacta con Placas Convectoras)", fontsize=16)
    plt.show()
```

```python
# =======================================================
# 4. VISUALIZACIÓN LIMPIA (Estilo Fig. 16 - Sin Líneas)
# =======================================================
if mapas_guardados:
    # Ajustamos la figura para que sea panorámica y las bobinas se vean con la proporción correcta
    fig, axes = plt.subplots(1, 4, figsize=(18, 9), constrained_layout=True, dpi=120)

    # Etiquetas según los tiempos capturados
    labels = ['(a) 5 h', '(b) 15 h', '(d) 20 h', '(e) 25 h']

    for idx, t_val in enumerate(TIEMPOS_OBJETIVO_H):
        if t_val in mapas_guardados and idx < 4:
            ax = axes[idx]
            data_C = mapas_guardados[t_val] - 273.15 # Convertir a Celsius

            # Rotar y voltear para que: Eje Y = Altura (Vertical), Eje X = Radio (Horizontal)
            plot_data = np.flipud(np.rot90(data_C))

            # --- AJUSTE DE COLOR (CONTRASTE) ---
            # Usamos percentiles para saturar los extremos y que se vea vívido
            vmin = np.percentile(data_C, 2)
            vmax = np.percentile(data_C, 98)

            # MAPA DE COLOR PURO (Sin contour)
            im = ax.imshow(plot_data, cmap='jet', aspect='auto', interpolation='bicubic',
                           extent=[0, 1, 0, NUM_COILS], vmin=vmin, vmax=vmax)

            # SEPARADORES DE BOBINAS (Placas convectoras)
            # Dibuja una línea blanca entre cada bobina para simular el stack
            for k in range(1, NUM_COILS):
                ax.axhline(y=k, color='white', linewidth=3, linestyle='-')

            # Títulos y limpieza de ejes
            ax.set_title(labels[idx], fontsize=16, fontname='serif', pad=12)
            ax.set_xticks([]) # Sin números en eje X
            ax.set_yticks([]) # Sin números en eje Y

            # Barra de color
            cbar = plt.colorbar(im, ax=ax, orientation='horizontal', pad=0.03, fraction=0.05)
            cbar.ax.tick_params(labelsize=10)
            if idx == 0:
                cbar.set_label('Temperature (°C)', fontsize=11)

    plt.suptitle("Temperature Field Evolution (Clean View)", fontsize=18, y=1.05)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =======================================================
# 1. PARÁMETROS DEL MODELO (CALIBRADO PARA ALTA T)
# =======================================================
# Geometría
Diam_ext = 1.867
Diam_int = 0.508
Width = 1.250
NUM_COILS = 4
H_total = Width * NUM_COILS

# Mallado
M = 20
N_coil = 30
N = N_coil * NUM_COILS
dr = (Diam_ext - Diam_int) / 2 / (M - 1)
dz = Width / (N_coil - 1)
r_coords = np.linspace(Diam_int/2, Diam_ext/2, M)

dt = 2.0

# Propiedades
rho = 7850.0; c_p = 500.0; lambda_z = 45.0; lambda_s = 45.0
S_strip = 0.00150; b_o = 10.0e-6; lambda_o = 1.5; epsilon = 0.08554 # Emisividad baja del acero
sigma_sb = 5.6703e-8; P_mpa = 8.0; H_mpa = 1133.86
sigma_p = 3.2215e-6; tan_theta = 0.1

# Gas y Flujo (AJUSTADO)
u_gas_base = 20.0 # Aumentado de 15 a 20 m/s para mejor convección base
factors_flow_radial = [1.0, 0.798, 0.652, 0.543]
factors_flow_axial_top = [0.202, 0.183, 0.167, 0.543]

# Factor de Radiación (Eq. 8 del Paper)
# El paper usa un "Composite heat transfer coefficient" para T > 800K.
# Este factor 'k' simula la radiación de la cubierta interior hacia la bobina.
K_RADIATION_FACTOR = 3.0

# Tiempos
t_subida = 5.0 * 3600
t_mant = 12.5 * 3600
t_bajada = 10.0 * 3600
total_steps = int((t_subida + t_mant + t_bajada) / dt)

TIEMPOS_OBJETIVO_H = [5.0, 15.0, 18.0, 20.0, 25.0]

# =======================================================
# 2. FUNCIONES FÍSICAS (CON RADIACIÓN)
# =======================================================
def props_gas_H2(T):
    T = np.maximum(T, 100.0)
    lam_g = 0.18 + 3e-4 * (T - 300)
    mu_g = 8.9e-6 + 2e-8 * (T - 300)
    return lam_g, mu_g

def calc_h_composite(T_gas, u_gas, flow_ratio):
    """Calcula h convectivo Y aplica el factor de radiación si T > 800K"""
    if flow_ratio <= 0.001: return 0.0

    # 1. Convección Pura (Dittus-Boelter)
    lam_g, mu_g = props_gas_H2(T_gas)
    rho_g = 0.08 * (300/T_gas)
    De = 0.1
    u_eff = u_gas * flow_ratio
    Re = np.maximum((rho_g * u_eff * De) / mu_g, 100.0)
    Nu = 0.023 * (Re**0.8) * (0.7**0.4)
    h_conv = (Nu * lam_g) / De

    # 2. Mejora por Radiación (Eq. 8 Paper)
    # Si el gas (y por ende el entorno/horno) está muy caliente,
    # la radiación domina. Multiplicamos h por un factor k.
    if T_gas >= 800.0:
        return h_conv * K_RADIATION_FACTOR
    else:
        # Transición suave opcional, o paso directo como dice el paper
        return h_conv

def calc_lambda_r(T_node):
    P = P_mpa * 1e6; H_ = H_mpa * 1e6
    phi = P / (H_ + P)
    b = 42.7e-6 * np.exp(-5e-2 * P_mpa)
    lam_g, _ = props_gas_H2(T_node)
    R_s2 = S_strip/(2*lambda_s); R_o = b_o/lambda_o
    T_k = np.maximum(T_node, 100.0)
    R_r = (2-epsilon)/(4*(1-phi)*epsilon*sigma_sb*(T_k**3))
    R_g = b/((1-phi)*lam_g)
    R_d = (sigma_p/(1.13*lambda_s*tan_theta))*(phi**-0.94)
    R_gap = 1.0/((1/R_r)+(1/R_g)+(1/R_d))
    return S_strip / (2*R_s2 + 2*R_o + R_gap)

def get_T_furnace(t_sec):
    if t_sec <= t_subida:
        return 300.0 + (983.0 - 300.0)*(t_sec/t_subida)
    elif t_sec <= (t_subida + t_mant):
        return 983.0
    else:
        dt_cool = t_sec - (t_subida + t_mant)
        return 983.0 - (983.0 - 300.0)*(dt_cool/t_bajada)

# =======================================================
# 3. SIMULACIÓN
# =======================================================
T = np.full((M, N), 300.0)
T_new = T.copy()
mapas_guardados = {}
C1 = dt / (rho * c_p)

print("Simulando con Factor de Radiación (Objetivo: ~680°C)...")

for step in range(total_steps):
    t_curr = step * dt
    T_gas_furnace = get_T_furnace(t_curr)

    # Modelo de gas en cascada (Menos pérdida para mantener calor)
    T_gas_coils = np.zeros(NUM_COILS)
    tg = T_gas_furnace
    for k in range(NUM_COILS):
        T_gas_coils[k] = tg
        T_surf = np.mean(T[:, k*N_coil:(k+1)*N_coil])
        # Factor de caída reducido a 0.02 (el gas mantiene mejor su temp)
        tg = tg - 0.02 * (tg - T_surf)

    # Lambda_r field
    L_r_field = np.zeros((M, N))
    for i in range(M):
        for j in range(N):
            L_r_field[i,j] = calc_lambda_r(T[i,j])

    # FDM Interno
    for i in range(1, M-1):
        r_i = r_coords[i]
        term_common = C1 * L_r_field[i,:]
        dT_dr2 = (T[i+1,:] - 2*T[i,:] + T[i-1,:]) / dr**2
        dT_dr  = (T[i+1,:] - T[i-1,:]) / (2*dr)
        radial = term_common * (dT_dr2 + (1/r_i)*dT_dr)
        dT_dz2 = (T[i, 2:] - 2*T[i, 1:-1] + T[i, :-2]) / dz**2
        axial = (C1 * lambda_z) * dT_dz2
        T_new[i, 1:-1] = T[i, 1:-1] + radial[1:-1] + axial

    # Condiciones de Contorno con H COMPOSITE (Radiación)
    for k in range(NUM_COILS):
        j_s = k * N_coil; j_e = (k + 1) * N_coil
        Tg = T_gas_coils[k]

        # Usamos la nueva función calc_h_composite
        h_rad = calc_h_composite(Tg, u_gas_base, factors_flow_radial[k])

        # Radial BC
        for j in range(j_s, j_e):
            Bi_in = h_rad * dr / L_r_field[1, j]
            T_new[0, j] = (T[1, j] + Bi_in * Tg) / (1 + Bi_in)
            Bi_out = h_rad * dr / L_r_field[M-1, j]
            T_new[M-1, j] = (T[M-2, j] + Bi_out * Tg) / (1 + Bi_out)

        # Axial BC
        ratio_bot = factors_flow_axial_top[k-1] if k > 0 else 0.2
        h_bot = calc_h_composite(Tg, u_gas_base, ratio_bot)
        Bi_bot = h_bot * dz / lambda_z

        ratio_top = factors_flow_axial_top[k]
        h_top = calc_h_composite(Tg, u_gas_base, ratio_top)
        Bi_top = h_top * dz / lambda_z

        for i in range(M):
            T_new[i, j_s] = (T[i, j_s+1] + Bi_bot * Tg) / (1 + Bi_bot)
            T_new[i, j_e-1] = (T[i, j_e-2] + Bi_top * Tg) / (1 + Bi_top)

    T = T_new.copy()

    # Guardar frames
    t_h = t_curr / 3600.0
    for t_obj in TIEMPOS_OBJETIVO_H:
        if abs(t_h - t_obj) < (dt/3600.0 * 2.0) and t_obj not in mapas_guardados:
            print(f"Capturado {t_obj} h - Max Temp actual: {np.max(T)-273.15:.1f} C")
            mapas_guardados[t_obj] = T.copy()

# =======================================================
# 4. VISUALIZACIÓN FINAL (5 PANELES - FIG 16)
# =======================================================
if len(mapas_guardados) >= 5:
    fig, axes = plt.subplots(1, 5, figsize=(20, 8), constrained_layout=True, dpi=120)

    labels = ['(a) 5 h', '(b) 15 h', '(c) 18 h', '(d) 20 h', '(e) 25 h']
    sorted_times = sorted(TIEMPOS_OBJETIVO_H)

    for idx, t_val in enumerate(sorted_times):
        ax = axes[idx]
        data_C = mapas_guardados[t_val] - 273.15

        plot_data = np.flipud(np.rot90(data_C))

        # Percentiles para contraste estilo paper
        vmin = np.percentile(data_C, 2)
        vmax = np.percentile(data_C, 98)

        im = ax.imshow(plot_data, cmap='jet', aspect='auto', interpolation='bicubic',
                       extent=[0, 1, 0, NUM_COILS], vmin=vmin, vmax=vmax)

        for k in range(1, NUM_COILS):
            ax.axhline(y=k, color='white', linewidth=2)

        ax.set_title(labels[idx], fontsize=14, fontname='serif', weight='bold')
        ax.set_xticks([]); ax.set_yticks([])

        cbar = plt.colorbar(im, ax=ax, orientation='vertical', fraction=0.046, pad=0.04)
        cbar.ax.tick_params(labelsize=9)

    plt.suptitle("Steel coil cross-section temperature field (Corrected Radiation Physics)", fontsize=16, y=1.05)
    plt.show()
else:
    print(f"Error: Solo se capturaron {len(mapas_guardados)} tiempos.")
```
